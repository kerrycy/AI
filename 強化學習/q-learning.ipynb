```python
# 設定一開始採用何種策略的參數theta_0

# 列為狀態0～23、欄移動方向的↑、→、↓、←
theta_0 = np.array([[np.nan, 1, 1, np.nan],  # s0
           [np.nan, 1, 1, 1],  # s1
           [np.nan, np.nan, np.nan, 1],  # s2
           [np.nan, 1, 1, np.nan],  # s3
           [np.nan, np.nan, 1, 1],  # s4
           [1, 1, 1, np.nan],  # s5
           [1, 1, 1, 1],  # s6
           [np.nan, 1, 1, 1],  # s7
           [1, 1, 1, 1],  # s8
           [1, np.nan, 1, 1],  # s9
           [1, 1, 1, np.nan],  # s10
           [1, 1, 1, 1],  # s11
           [1, 1, 1, 1],  # s12
           [1, np.nan, np.nan, 1],  # s13
           [1, np.nan, 1, np.nan],  # s14
           [1, 1, 1, np.nan],  # s15
           [np.nan, np.nan, 1, 1],  # s16
           [1, 1, 1, np.nan],  # s17
           [1, np.nan, np.nan, 1],  # s18
           [np.nan, np.nan, 1, np.nan],  # s19
           [1, 1, np.nan, np.nan],  # s20
           [1, 1, np.nan, 1],  # s21
           [1, 1, np.nan, 1],  # s22
           [np.nan, 1, np.nan, 1],  # s23
                    ])
```

```python
def softmax_convert_into_pi_from_there(theta):
  beta=1.0
  [m,n]=theta.shape
  pi=np.zeros((m,n))
  exp_theta=np.exp(beta * theta)
  for i in range(0,m):
      pi[i,:]=exp_theta[i,:] / np.nansum(exp_theta[i,:])
  pi=np.nan_to_num(pi) # 將nan轉換成0
  return pi
pi_0 = softmax_convert_into_pi_from_there(theta_0)
print(pi_0)
```

```python
def get_action_and_next_s(pi,s): # 設定智能體的下一步狀態
    direction=["up","right","down","left"]
    next_direction=np.random.choice(direction,p=pi[s,:])
    # 根據概率去選擇方向
    if next_direction=="up":
        action=0
        s_next=s-5 # 向上移動時狀態數字-5
    if next_direction=="right":
        action=1
        s_next=s+1 # 向右移動時狀態數字+1
    if next_direction=="down":
        action=2
        s_next=s+5 # 向下移動時狀態數字+5
    if next_direction=="left":
        action=3
        s_next=s-1 # 向左移動時狀態數字-1
    return [action,s_next]



```

```python
def goal_maze_ret_s_a(pi): # 根據定義的策略持續移動.
    s=0 # 設定開始地點
    s_a_history=[[0,np.nan]]  # 記錄智能體軌跡的串列
    while (1): # 回圈執行，直到智能體到達終點
        [action,next_s]=get_action_and_next_s(pi,s)
        s_a_history[-1][1]=action # 代入當前狀態 表示最后一個狀態的動作

        s_a_history.append([next_s,np.nan])
        # 代入下一個狀態，因為不知道他的動作，所以用nan
        if next_s==24:
            break
        else:
            s=next_s
    return s_a_history
```

```python
s_a_history = goal_maze_ret_s_a(pi_0)
print(s_a_history)
print("走出迷宮的總步數為" + str(len(s_a_history) - 1) + "嘿")
```

```python
def update_theta(theta,pi,s_a_history):
    eta=0.02  #學習率
    T=len(s_a_history)-1
    [m,n]=theta.shape
    delta_theta=theta.copy()

    for i in range(0,m):
        for j in range(0,n):
            if not(np.isnan(theta[i,j])):
                SA_i=[SA for SA in s_a_history if SA[0] == i]
                # 取狀態i
                SA_ij=[SA for SA in s_a_history if SA == [i,j]]
                # 取狀態i下應該採取的動作
                N_i=len(SA_i) # 狀態i下動作總次數
                N_ij=len(SA_ij) # 狀態i下採取動作j的次數
                delta_theta[i,j]=(N_ij-pi[i,j]*N_i)/T
    new_theta=theta+eta*delta_theta
    return new_theta
```

```python
new_theta = update_theta(theta_0, pi_0, s_a_history)
pi = softmax_convert_into_pi_from_there(new_theta)
print(pi)
```

```python
stop_epsilon=10**-4 
theta=theta_0
pi=pi_0
is_continue=True
count=1
while is_continue:
  s_a_history=goal_maze_ret_s_a(pi) # 由策略Π去搜索
  new_theta=update_theta(theta,pi,s_a_history) # 更新引數
  new_pi=softmax_convert_into_pi_from_there(new_theta) # 更新策略

  print(np.sum(np.abs(new_pi-pi))) # 輸出策略變化
  print("走出迷宮的步數為"+str(len(s_a_history)-1)+"步")
  if np.sum(np.abs(new_pi-pi))<stop_epsilon:
      is_continue=False
  else:
      theta=new_theta
      pi=new_pi
print(s_a_history)
```
